**6 факторів**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import FactorAnalysis
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(f)

X = StandardScaler().fit_transform(X_imputed)



factors = 6


fas = [
    ("FA no rotation", FactorAnalysis(n_components=factors)),
    ("FA varimax", FactorAnalysis(n_components=factors, rotation="varimax")),
    ("FA quartimax", FactorAnalysis(n_components=factors, rotation="quartimax"))
]


fig, axes = plt.subplots(ncols=len(fas), figsize=(20, 16))


for ax, (title, fa) in zip(axes, fas):
    fa.fit(X)

    factor_matrix = fa.components_.T

    im = ax.imshow(factor_matrix, cmap="RdBu_r", vmax=1, vmin=-1)
    for (i, j), z in np.ndenumerate(factor_matrix):
        ax.text(j, i, str(z.round(2)), ha="center", va="center", color="black" if -0.1 < z < 0.1 else "white")

    ax.set_yticks(np.arange(len(f.columns)))
    if ax.get_subplotspec().is_first_col():
        ax.set_yticklabels(f.columns)
    else:
        ax.set_yticklabels([])
    ax.set_title(title)
    ax.set_xticks(np.arange(factors))
    ax.set_xticklabels([f"Factor {i+1}" for i in range(factors)])

cb = fig.colorbar(im, ax=axes, location='right', label="loadings")

#plt.savefig('load6fact.png')

plt.show()
```

```python
from sklearn.impute import SimpleImputer
from factor_analyzer import FactorAnalyzer
datasets=[f1,f2]
rotation=[None,'varimax','quartimax']
imputer = SimpleImputer(strategy='mean')
for df in datasets:

  X_imputed = imputer.fit_transform(df)
  X1 = StandardScaler().fit_transform(X_imputed)
  for rotat in rotation:
    print(rotat)
    fa1 = FactorAnalysis(n_components=5, rotation=rotat)
    fa1.fit(X1)
    factor_matrix = fa1.components_.T
    squared_loadings= np.square(factor_matrix)
    cov_matrix = np.cov(X1, rowvar=False)
    eigenvalues = np.linalg.eigvals(cov_matrix)
    eigenvalues=sorted(eigenvalues)
    factor1= squared_loadings[:, 0]
    factor2= squared_loadings[:, 1]
    factor3= squared_loadings[:, 2]
    factor4= squared_loadings[:, 3]
    factor5= squared_loadings[:, 4]
    factor1=factor1/eigenvalues[0]
    factor2=factor2/eigenvalues[1]
    factor3=factor3/eigenvalues[2]
    factor4=factor4/eigenvalues[3]
    factor5=factor5/eigenvalues[4]

    explained_variance = eigenvalues / np.sum(eigenvalues)

    factor1=factor1*explained_variance[0]
    factor2=factor2*explained_variance[1]
    factor3=factor3*explained_variance[2]
    factor4=factor4*explained_variance[3]
    factor5=factor5*explained_variance[4]
    print(sum(factor1)+sum(factor2)+sum(factor3)+sum(factor4)+sum(factor5))
```
